```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python

dataset.columns
```

```python
inde=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
inde= sc.fit_transform(inde)
inde= sc.transform(inde)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid={'criterion':['gini','entropy','log_loss'],'max_features':[None,'sqrt','log2'],'n_estimators':[10,50,100]}
grid=GridSearchCV(RandomForestClassifier(),param_grid,verbose=1,n_jobs=-1,refit=True,cv=5)
grid.fit(inde,dep)
```

```python
re=grid.cv_results_
print("best_parameters:",format(grid.best_params_))
grid_predictions=grid.predict(inde)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(dep,grid_predictions,average="weighted")
print("the the best f1_scores parameter{}:",format(grid.best_params_)),f1_score
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep,grid_predictions)
```

```python
print("confusion matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(dep,grid_predictions)
```

```python

print("claf report:\n",clf_report)
```
