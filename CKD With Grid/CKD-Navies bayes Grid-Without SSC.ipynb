```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python

dataset.columns
```

```python
inde=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.naive_bayes import MultinomialNB

from sklearn.model_selection import GridSearchCV



param_grid = {
    'alpha': [0.01, 0.1, 1.0, 2.0, 5.0],  # Smoothing parameter
    'fit_prior': [True, False]            # Whether to learn class prior probabilities
}
grid_search = GridSearchCV(estimator=MultinomialNB(), param_grid=param_grid, cv=5)

# Fit the model
grid_search.fit(inde, dep)

```

```python
re=grid_search.cv_results_
print("best_parameters:",format(grid_search.best_params_))
grid_search_predictions=grid_search.predict(inde)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(dep,grid_search.predict_proba(inde)[:,1])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep,grid_search_predictions)
```

```python
print("confusion matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(dep,grid_search_predictions)
```

```python

print("claf report:\n",clf_report)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.naive_bayes import ComplementNB

from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha': [0.01, 0.1, 1.0, 2.0, 5.0],  # Smoothing parameter
    'fit_prior': [True, False],           # Whether to learn class prior probabilities
    'norm': [True, False]                  # Whether to normalize the predicted probabilities
}
grid_search = GridSearchCV(estimator=ComplementNB(), param_grid=param_grid, cv=5)

# Fit the model
grid_search.fit(inde,dep)
```
