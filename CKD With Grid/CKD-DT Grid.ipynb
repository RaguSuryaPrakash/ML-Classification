```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python

dataset.columns
```

```python
inde=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Correct parameter grid for Decision Tree Classifier
param_grid = {
    'criterion': ['gini', 'entropy'],  # Options for classification
    'max_depth': [None, 5, 10, 15, 20],  # Optional max depth
    'min_samples_split': [2, 5, 10],      # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],        # Minimum samples at a leaf node
    'splitter': ['best', 'random'],        # Strategy for splitting nodes
}

# Create the GridSearchCV object
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1,cv=5)

# Fit the model
grid.fit(inde, dep)

# Best parameters
print("Best Parameters:", grid.best_params_)
```

```python
re=grid.cv_results_
print("best_parameters:",format(grid.best_params_))
grid_predictions=grid.predict(inde)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(dep,grid_predictions,average="weighted")
print("the the best f1_scores parameter{}:",format(grid.best_params_)),f1_score
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep,grid_predictions)
```

```python
print("confusion matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(dep,grid_predictions)
```
