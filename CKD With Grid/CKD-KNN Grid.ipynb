```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python

dataset.columns
```

```python
inde=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
inde= sc.fit_transform(inde)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(inde,dep)
```

```python
re=grid.cv_results_
print("best_parameters:",format(grid.best_params_))
grid_predictions=grid.predict(inde)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(dep,grid.predict_proba(inde)[:,1])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep,grid_predictions)
```

```python
print("confusion matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(dep,grid_predictions)
```

```python

print("claf report:\n",clf_report)
```
